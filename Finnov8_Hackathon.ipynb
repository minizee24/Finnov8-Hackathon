```python
# lets select compant name and industry features from the dataset
# lets select office country feature from the dataset
# lets select num_funding_rounds and total_funding_usd from the dataset

new_data = data[['company_name', 'industry', 'overview', 'office_region', 'num_funding_rounds', 'total_funding_usd', 'num_milestones', 'acquired', 'ipo', 'exit']]
```

```python
new_data
```

```python
new_data.isnull().sum()
```

```python
m_data = data.copy()
not_null_mask = m_data['industry'].notnull()
not_null_rows_1 = m_data[not_null_mask]
not_null_rows_1
```

```python
not_null_mask = not_null_rows_1['overview'].notnull()
not_null_rows_2 = not_null_rows_1[not_null_mask]

not_null_rows_2
```

```python
null_mask = m_data['industry'].isnull()
null_rows_1 = m_data[null_mask]
null_rows_1
```

```python
null_mask = null_rows_1['overview'].notnull()
null_rows_2 = null_rows_1[null_mask]
null_rows_2
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

company_overviews = not_null_rows_2['overview'].values.tolist()

industry_labels = not_null_rows_2['industry'].values.tolist()

# Train a classifier pipeline
classifier = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', LogisticRegression(n_jobs=-1)),
])

# Fit the classifier on the training data
classifier.fit(company_overviews, industry_labels)
```

```python
# Predict industry name using the trained classifier
predicted_industry = classifier.predict(null_rows_2['overview'])

# Print predicted industry name
predicted_industry
```

```python
null_rows_2['industry'] = predicted_industry
null_rows_2 # predicted_industry
```

```python
new_data.dropna(subset = ['industry'], inplace=True)
```

```python
new_data.dropna(subset = ['overview'], inplace=True)
```

```python
new_data
```

```python
p_data = null_rows_2[['company_name', 'industry', 'overview', 'office_region', 'num_funding_rounds', 'total_funding_usd', 'num_milestones', 'acquired', 'ipo', 'exit']]
```

```python
p_data
```

```python
final_data = new_data.append(p_data, ignore_index=True)
```

```python
final_data
```

```python
final_data = final_data.drop(columns=['overview'])
final_data = final_data.drop(columns=['company_name'])

#final_data["company_name"]= pd.factorize(final_data.company_name)[0]

final_data['industry']= pd.factorize(final_data.industry)[0]
final_data['office_region']= pd.factorize(final_data.office_region)[0]
```

```python
final_data
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_data.iloc[:,:5], final_data.iloc[:,5:], test_size=0.25, random_state=1)
```

```python
# Initialize the base classifier (CatBoostClassifier in this example)
base_classifier = CatBoostClassifier(iterations = 1000, depth = 4)

# Initialize the MultiOutputClassifier with the base classifier
multi_target_classifier = MultiOutputClassifier(base_classifier, n_jobs=-1)

# Train the multi-output classifier
multi_target_classifier.fit(X_train, y_train)

# Predict on the testing set
y_pred = multi_target_classifier.predict(X_test)

# Evaluate performance
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100)
```
